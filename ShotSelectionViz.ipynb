```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('C:/Users/William Jiang/Documents/tennis_MatchChartingProject/charting-m-stats-ShotDirOutcomes_1.csv', engine='python')
```

```python
df[['Date','Gender','City','Round','P1Name','P2Name']] = df['match_id'].str.split(expand=True,pat = "-")
```

```python
df[['ForB','ShotDirection']] = df['row'].str.split(expand=True,pat = "-")
```

```python
def playermadeshot(row):
    if int(row['player'])==1:
        return_val= row['P1Name']
    else:
        return_val= row['P2Name']
    return return_val
```

```python
def playerreceivingshot(row):
    if int(row['player'])==2:
        return_val= row['P1Name']
    else:
        return_val= row['P2Name']
    return return_val
```

```python
df['playermadeshot']=df.apply(playermadeshot,axis=1)
df['playerreceivingshot']=df.apply(playerreceivingshot,axis=1)
```

```python
df[['Date','Gender','City','Round','P1Name','P2Name','playermadeshot','playerreceivingshot']]=df[['Date','Gender','City','Round','P1Name','P2Name','playermadeshot','playerreceivingshot']].replace('_', ' ', regex=True)
```

```python
df['Year']=df['Date'].str.slice(stop=4)
```
